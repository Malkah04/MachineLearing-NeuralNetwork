```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv("dataa.csv")

data = pd.DataFrame(data)

data.plot(x='CostPaid' ,y="SpeedOfService" ,kind='scatter')
# داتا زي الزفت مش هينفع اشتغل عليها اصلا هصوووووت
```

```python
x= data[["CostPaid"]]  # to be 2D
y= data["SpeedOfService"] # 1D
         

```

```python

x_train ,x_test ,y_train,y_test =train_test_split(x ,y,test_size=0.2 ,random_state=0)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python

model =LinearRegression()

```

```python

model.fit(x_train ,y_train)

```

```python
y_predict =model.predict(x_test)
```

```python
error =mean_squared_error(y_test ,y_predict)
print(error)
```

```python
plt.scatter(x_train ,y_train ,color="green" ,label="training data")
plt.scatter(x_test ,y_test ,color="blue" ,label="test data")
plt.plot(x_test , y_predict ,color="red" , label="regression line")

plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.title("LinearRegression example 2")
plt.show()
```
