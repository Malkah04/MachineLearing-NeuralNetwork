```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.preprocessing import PolynomialFeatures
```

```python
data =pd.read_csv("Diabetes Classification.csv")
print(data.head())

print(pd.value_counts(data["Blood Pressure"]))

print(pd.value_counts(data["Family History of Diabetes"]))

print(pd.value_counts(data["Diet"]))


```

```python
data["Gender"] =data["Gender"].map({ "Female":0 , "Male":1 })
data["Pressure"] =data["Blood Pressure"].map({ "Normal":0 , "High":1 ,"Low":2 })
data["Smoking"] =data["Smoking"].map({ "Yes":0 , "No":1 })

data["Diet"] =data["Diet"].map({ "Poor":0 , "Healthy":1 })
data["Diagnosis"] =data["Diet"].map({ "No":0 , "Yes":1 })
data["Exercise"] =data["Diet"].map({ "No":0 , "Yes":1 })



data["Family History of Diabetes"]=data["Family History of Diabetes"].map({"No":0 ,"Yes":1 })



```

```python
X = data.drop(["Diagnosis"],axis=1)
Y =data["Diagnosis"]

x_train,x_test ,y_train ,y_test =train_test_split(X,Y ,train_size=0.2 ,random_state=42)
```

```python
model =LogisticRegression(max_iter=10000)

model.fit(x_train ,y_train)

y_pred =model.predict(x_test)

print("accuracy" ,accuracy_score(y_test ,y_pred))

print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
